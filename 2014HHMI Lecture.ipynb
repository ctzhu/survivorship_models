<h1 align="left", style="color:green">2. Basic concepts</h1>

$$t\Huge$$  
===================
<h4 align="center"> Death time, Event time,....</h4> 

```python
t_pcb_f = [74, 485, 520, 599, 699, 713, 719, 722, 724, 733, 817, 824, 855, 857, 
         870, 984, 997, 1023, 1039, 1074, 1143, 1189]
t_dr_f  = [48, 62, 259, 260, 296, 342, 410, 755, 887, 901, 904, 925, 938, 980, 
         999, 1041, 1107, 1225, 1232, 1261, 1304, 1421]
t_pcb_f = np.array(t_pcb_f, dtype=float) #the death days for ab libitum female
t_dr_f  = np.array(t_dr_f, dtype=float) #the death days for dietary restriction female
```

$$S(t); 1-S(t)=F(t)\Huge$$  
===================
<h4 align="center"> Survivorship; Cumulative Distribution Function </h4> 

```python
d_pcb_f = (np.arange(len(t_pcb_f))+1.0)/len(t_pcb_f)
d_dr_f  = (np.arange(len(t_dr_f))+1.0)/len(t_dr_f)
s_pcb_f = 1-d_pcb_f
s_dr_f = 1-d_dr_f
print s_pcb_f #the survivorship over time
```

$$H(t)={S}'(t)/S(t)=dS/dt\cdot1/S(t)\Huge$$  
===================
<h4 align="center"> Hazard function, Instantaneous death rate,....</h4> 

```python
h_pcb_f = np.diff(d_pcb_f)/s_pcb_f[:-1]/np.diff(t_pcb_f)
h_dr_f = np.diff(d_dr_f)/s_dr_f[:-1]/np.diff(t_dr_f)
print h_pcb_f
print np.diff(d_pcb_f)/d_pcb_f[:-1]
```

<h1 align="left", style="color:green">4. Basic Models</h1>

## $$H(t)=C; S(t)=e^{-C\cdot t}\Huge$$   Exponential Decay <p><p>  $$H(t)=(\kappa/\lambda)\cdot(t/\lambda)^{\kappa-1}\Huge$$ $$log(H(t))=log(\kappa \cdot \lambda^{-\kappa})+(\kappa-1)\cdot log(x)\Huge$$  Weibull model (Mortality increase with time in a power function relationship) <p><p>  $$H(t)=\alpha e^{\beta t}$$ $$log(H(t))=log(\alpha)+ \beta t \Huge$$   Gompertz model (Mortality increase with time Exponentially)

```python
import scipy.stats as ss
t=np.linspace(1, 1400, 100)
def hzd(func, x, p):
    return func.pdf(x, *p)/func.sf(x, *p)
plt.plot(t_pcb_f[:-1], np.log(h_pcb_f), 'o', label='ab libitum', markersize=10, alpha=0.6)
plt.plot(t, np.log(hzd(ss.expon, t, ss.expon.fit(t_pcb_f, floc=0))), label='Exponential')
plt.plot(t, np.log(hzd(ss.weibull_min, t, ss.weibull_min.fit(t_pcb_f, floc=0))), 
         label='Weibull')
plt.plot(t, np.log(hzd(ss.gompertz, t, ss.gompertz.fit(t_pcb_f, floc=0))), label='Gompertz')
plt.ylim((-10,-2))
plt.legend(loc='upper left')
_=plt.xlim(xmax=1450)
```

```python
plt.plot(np.log(t_pcb_f[:-1]), np.log(h_pcb_f), 'o', label='ab libitum', markersize=10, alpha=0.6)
plt.plot(np.log(t), np.log(hzd(ss.expon, t, ss.expon.fit(t_pcb_f, floc=0))), label='Exponential')
plt.plot(np.log(t), np.log(hzd(ss.weibull_min, t, ss.weibull_min.fit(t_pcb_f, floc=0))), 
         label='Weibull')
plt.plot(np.log(t), np.log(hzd(ss.gompertz, t, ss.gompertz.fit(t_pcb_f, floc=0))), label='Gompertz')
plt.ylim((-10, -2))
plt.xlim((4, np.log(1450)))
_=plt.legend(loc='upper left')
```

<h1 align="left", style="color:green">5. Mixture Models</h1>

## Mixed Mortality, combination of age-dependent -independent source <p><p>  $$H(t)=(\kappa/\lambda)\cdot(x/\lambda)^{\kappa-1}+C\Huge$$   Weibull model with extrinsic mortality <p><p>  $$H(t)=\alpha e^{\beta t}+C\Huge$$   Gompertz model with extrinsic mortality (Gompertz-Makehamm Model)

```python
dr_k, dr_l, dr_cw   =6.97453798004, 1195.22639259, 0.000653408897982
dr_a, dr_b, dr_cg   =5.93236884108e-06, 0.00571527646874, 0.00063822020057
pcb_k, pcb_l, pcb_cw=5.12354741343, 926.947225217, 0.000122215890997
pcb_a, pcb_b, pcb_cg=4.14104391332e-05, 0.00521423796061, 6.3581886375e-05
weibull_mixed =lambda x, k, l, c: c+(k/l)*((x/l)**(k-1))
gompertz_mixed=lambda x, a, b, c: c+a*np.exp(b*x)

plt.plot(t_pcb_f[:-1], np.log(h_pcb_f), 'o', label='ab libitum', markersize=10, alpha=0.6)
plt.plot(t, np.log(weibull_mixed(t, pcb_k, pcb_l, pcb_cw)), 'b--', label='Weibull_mixed')
plt.plot(t, np.log(gompertz_mixed(t, pcb_a, pcb_b, pcb_cg)), 'b-', label='Gompertz_mixed')

plt.plot(t_dr_f[:-1], np.log(h_dr_f), 'go', label='ab libitum', markersize=10, alpha=0.6)
plt.plot(t, np.log(weibull_mixed(t, dr_k, dr_l, dr_cw)), 'g--', label='Weibull_mixed')
plt.plot(t, np.log(gompertz_mixed(t, dr_a, dr_b, dr_cg)), 'g-', label='Gompertz_mixed')

plt.ylim(ymin=-10)
_=plt.xlim(xmax=1450)
plt.legend(loc='upper left')
```

## Fit to survivalship, with extrinsic mortality also modelled

```python
import scipy.integrate as si
s_weibull_mixed = lambda x, k, l, c: np.exp(-si.quad(weibull_mixed, 1e-5, x, args=(k, l, c))[0])
s_gompertz_mixed = lambda x, a, b, c: np.exp(-si.quad(gompertz_mixed, 1e-5, x, args=(a, b, c))[0])

plt.plot(t_pcb_f, s_pcb_f, '.-', label='ab libitum')
plt.plot(t_dr_f, s_dr_f, '.-', label='DR')
plt.legend()
plt.vlines(t_pcb_f.mean(), ymin=0, ymax=1, linestyle='--', color='b', label='ab libitum')
plt.vlines(t_dr_f.mean(), ymin=0, ymax=1, linestyle='--', color='g', label='DR')
plt.plot(t, [s_gompertz_mixed(item, pcb_a, pcb_b, pcb_cg) for item in t], linestyle='--', color='b')
plt.plot(t, [s_gompertz_mixed(item, dr_a, dr_b, dr_cg) for item in t], linestyle='--', color='g')
plt.hlines(0.5, 0, xmax=plt.xlim()[-1], linestyle=':')
_=plt.xlim(xmax=1450)
```
